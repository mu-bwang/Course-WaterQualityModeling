# Example 14.1 point estimation of river hydrogeometry

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([0,4,8,12,16,20])
z = np.array([0,0.4,1,1.5,0.2,0])
U0_2 = np.array([0, np.nan, 0.2, 0.3, np.nan, 0])
U0_6 = np.array([0, 0.05, np.nan, np.nan, 0.07, 0])
U0_8 = np.array([0, np.nan, 0.12, 0.2, np.nan, 0])
```

## Determine depth averaged velocity at each measurement location

```python
# stack velocity measurement into a 2D array
Ux = np.vstack((U0_2, U0_6, U0_8))

# calculate depth averaged velocity at each measurement point
Um = np.nanmean(Ux, axis=0)
```

```python
plt.plot(x,-z,'o-',label='bed profile (m)')
plt.plot(x,Um,'s-',label='depth-avg velocity (m/s)')
plt.xlabel('x (m)')
plt.legend()
```

## Find the cross section area uisng trapezoidal rule

```python
Ac = np.trapz(z, x)
print('cross section area = ', Ac, 'm^2')
```

```python
H = Ac/max(x)
print('mean water depth = ', H, 'm')
```

## Estimate flow discharge and mean velocity

```python
Q = np.trapz(z*Um, x)
print('discharge = ', Q, 'm^3/s')
```

```python
U_mean = Q/Ac
print('mean velocity = ', U_mean, 'm/s')
```
