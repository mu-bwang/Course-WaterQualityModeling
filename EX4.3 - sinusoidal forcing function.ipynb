Solution for the Example 4.3 in Chapra book: Surface water-quality modeling

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
Tp = np.array([10, 1, 0.1])    # forcing period, yr
Q = 9e6       # inflow = outflow, m^3/year
V = 2.5e6     # Lake volume, m^3
```

```python
eigenvalue = Q/V   # 1/year
angular_freq = 2*np.pi / Tp # 1/year
phase_shift_in_radian = np.arctan(angular_freq/eigenvalue)
phase_shift_in_year = phase_shift_in_radian * (Tp / (2*np.pi))
attenuation_coeff = eigenvalue / np.sqrt(eigenvalue**2 + angular_freq**2)
```

```python
print('attenuation coefficient = ', attenuation_coeff)
print('phase shift =', phase_shift_in_year * 365, 'days')
```

Now, let's look at the attenuation coefficient and phase shift as a function of loading period (or frequency). This relationship can be visualized by a figure called 'Bode diagrams'.

```python
Tp = np.logspace(-2,2)
angular_freq = 2*np.pi / Tp # 1/year
phase_shift_in_radian = np.arctan(angular_freq/eigenvalue)
phase_shift_in_year = phase_shift_in_radian * (Tp / (2*np.pi))
attenuation_coeff = eigenvalue / np.sqrt(eigenvalue**2 + angular_freq**2)
```

```python
fig, [ax1, ax2] = plt.subplots(nrows = 2, ncols = 1, figsize = (10,6))
ax1.plot(Tp, attenuation_coeff)
ax1.set_xscale('log')
ax1.grid("True")
ax1.set_ylabel('Amplitude reduction')
ax2.plot(Tp, np.rad2deg(phase_shift_in_radian))
ax2.set_xscale('log')
ax2.grid("True")
ax2.set_xlabel('Period of forcing function (yr)')
ax2.set_ylabel('Phase shift (degrees)')
```
