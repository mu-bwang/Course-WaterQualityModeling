Compare residence time between CSTR and PFR as a function of performance eficiency 

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
beta = np.linspace(0.01,1)

# reaction rate does not matter when calculating the ratio of residence time
k = 1
t_CSTR = 1/k * (1-beta)/beta
t_PFR = 1/k * np.log(1/beta)

plt.plot(beta, t_CSTR/t_PFR)
plt.plot(beta, np.ones(len(beta)),'--')
plt.grid(True)
plt.xlabel(r'$\beta$',fontsize=20)
plt.ylabel(r'$\tau_{CSTR}/\tau_{PFR}$',fontsize=20)
```
