```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt

matplotlib.rc('xtick', labelsize=20) 
matplotlib.rc('ytick', labelsize=20) 
matplotlib.rc('axes', labelsize=20) 
```

```python
# dimensionless time 
t = np.linspace(0,10,num=100)

# dimensioless reaction rate
k = np.array([0.2, 0.5, 1, 2, 5])
```

```python
ca = np.exp(-t)
plt.figure(figsize=(12, 8))
plt.plot(t,ca,'b',label='A')
for ki in k:
    cb = 1/(1-ki) * (np.exp(-ki*t) - np.exp(-t))
    cc = 1-ca-cb
    
    plt.plot(t,cb,label = 'B, k*='+ str(ki))
    plt.plot(t,cc,'--', label = 'C, k*='+ str(ki))
    plt.legend()
    
plt.xlabel('Dimensionless time, t')
plt.ylabel('Dimensionless concentration, c')
```
